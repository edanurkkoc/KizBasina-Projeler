```python
print(df.columns.tolist())

```

```python
traffic_by_year = df.groupby("AADFYear")["AllMotorVehicles"].sum().reset_index()

sns.lineplot(data=traffic_by_year, x="AADFYear", y="AllMotorVehicles", marker="o")
plt.title("Yıllara Göre Toplam Trafik Hacmi (UK)")
plt.xlabel("Yıl")
plt.ylabel("Toplam Araç Sayısı")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
road_traffic = df.groupby("RoadCategory")["AllMotorVehicles"].sum().sort_values()

sns.barplot(x=road_traffic.values, y=road_traffic.index, palette="viridis")
plt.title("Yol Türlerine Göre Trafik Hacmi")
plt.xlabel("Toplam Araç")
plt.ylabel("Yol Türü")
plt.show()
```

```python
vehicle_columns = ['PedalCycles', 'Motorcycles', 'CarsTaxis', 'BusesCoaches',
                   'LightGoodsVehicles', 'AllHGVs']

vehicle_totals = df[vehicle_columns].sum().sort_values()

sns.barplot(x=vehicle_totals.values, y=vehicle_totals.index, palette="rocket")
plt.title("Araç Türlerine Göre Toplam Trafik")
plt.xlabel("Araç Sayısı")
plt.ylabel("Araç Türü")
plt.show()
```

```python
region_traffic = df.groupby("Region")["AllMotorVehicles"].sum().sort_values(ascending=False)

sns.barplot(x=region_traffic.values, y=region_traffic.index, palette="magma")
plt.title("Bölgelere Göre Toplam Trafik")
plt.xlabel("Araç Sayısı")
plt.ylabel("Bölge")
plt.show()
```
